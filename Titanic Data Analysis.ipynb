## TRAIN DATA

```python
X= train_data.drop('Survived',axis=1)
y= train_data['Survived']
```

```python
from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

LogM=LogisticRegression()
```

```python
LogM.fit(X_train,y_train)
```

```python
predictions=LogM.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```

## SAVING MODEL

```python
import pickle

pickle.dump(LogM,open("Logistic_Model_first_pkl","wb"))
```

```python
# Load a saved pickle model
loaded_pickle_model = pickle.load(open("Logistic_Model_first_pkl", "rb"))

# Evaluate loaded model
loaded_pickle_model.score(X_test, y_test)
```

```python
# Saving a model with joblib
from joblib import dump, load

# Save a model to file
dump(LogM ,filename="gs_Logistic_regression_model_1.joblib") 
```
